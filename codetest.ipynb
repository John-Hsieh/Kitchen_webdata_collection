````python
#è·å–ç½‘å€
import requests
from bs4 import BeautifulSoup
url = "https://www.xiachufang.com/category/40076/"
r = requests.get(url)
htmls = r.text
print(htmls)
soup = BeautifulSoup(htmls, 'html.parser')            #lxmlä¹Ÿå¯ä»¥ç”¨
````

````python
import json
#æ•´åˆåçš„ç‰ˆæœ¬

def soupprocessor(soup, number):
    scores = soup.find_all(class_="stats",recursive=True)
    titles = soup.find_all(class_="name", recursive=True)

    scorecount = 0
    titlecount = 0
    #è®¡ç®—ç¬¬ä¸€éƒ¨åˆ†çš„å€¼
    titleholder = []
    scoreholder = []

    for i in titles:
        if(i.find(target="_blank") is None):
            continue;
        else:
            #print(i.find(target="_blank").get_text().strip( ))
            titleholder.append(i.find(target="_blank").get_text().strip( ))
            titlecount = titlecount+1


    #è®¡ç®—ç¬¬äºŒéƒ¨åˆ†çš„å€¼
    for i in scores:
        if(i.find(class_ = "score bold green-font") is None):
            #print("Not scored")
            scoreholder.append("null")
        elif(type(i.find(class_ = "score bold green-font")).__name__ == 'Tag'):
            #print(i.find(class_ = "score bold green-font").get_text())
            scoreholder.append(i.find(class_ = "score bold green-font").get_text())
        else:
            print("Unexpected condition")
        scorecount = scorecount+1


    print("score count:" + str(scorecount) + " and " + "title count: " + str(titlecount))
    if(scorecount != titlecount):
        print("item number do not match, error occured")
    else:
        print("item count && score count match, perform writing")
    
    print(scorecount)
    for i in range(0,scorecount-1):
        content.append({'recipe_title':titleholder[i],'recipe_rate':scoreholder[i]})
    #å¯¼å‡ºä¸ºjsonæ ¼å¼    



def jsonwrite():
    filename = "result1.json"#str(number)+".json"
    global content
    #if number == 1:
    #    file = open(filename,'w',encoding='utf-8')
    
    file = open(filename,'w+',encoding='utf-8')
    #content = json.dumps(content,ensure_ascii=False)
    #info = json.load(file)
    #info = info.append(content)
    #print(info)
    #file2 = open(filename,'w',encoding='utf-8')
    json.dump(content,file)

    file = open(filename,'r+',encoding='utf-8')
    info = json.load(file)
    print(info)
    #info = info.append(content)
    

//test content    

content = []
soupprocessor(soup, 1)
print(content)
jsonwrite()
````
